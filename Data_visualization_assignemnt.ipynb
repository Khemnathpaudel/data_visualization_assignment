```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap?
```

```python
np.random.seed(42)# for reproducibility

x1=np.random.normal(0,1,100)# variable x1,x2,x3 correlated with one another
x2=0.5*x1+np.random.normal(0,1,100)
x3=0.5*x1+0.5*x2 + np.random.normal(0,1,100)

df=pd.DataFrame({'x1':x1,'x2':x2,'x3':x3})

corr_matrix=df.corr()

sns.heatmap(corr_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
from bokeh.models import ColumnDataSource
output_notebook()

#generaating the dataset
np.random.seed(42)
x=np.random.rand(100)*100
y=np.random.rand(100)*100
sizes=np.random.rand(100)*30  #marker sizes
colors=np.random.randint(1,300,100) #markder colors

data = {
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
}
df = pd.DataFrame(data)

# create a columndatasource
source=ColumnDataSource(df)

p=figure(title='Bokeh scatter plot',x_axis_label="X",y_axis_label="Y")
p.scatter('x','y',size='sizes',color='colors',source=source)

show(p)

```

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Enable Bokeh to display plots in the notebook
output_notebook()

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create the scatter plot
p = figure(title="Bokeh Scatter Plot with Different Sizes and Colors",
           x_axis_label='X', y_axis_label='Y')

p.scatter('x', 'y', size='sizes', color='colors', alpha=0.6, source=source)

# Show the plot
show(p)
```

```python
#3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
```

```python
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
output_notebook()
#dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

data={'fruits':fruits,'counts':counts}

source=ColumnDataSource(data=data)

p=figure(x_range=fruits,title='fruit counts',x_axis_label='fruit',y_axis_label='count')

p.vbar(x='fruits',top='counts',width=0.5,source=source)

show(p)
```

```python
#4. Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
```

```python
output_notebook()
# Create a ColumnDataSource
source = ColumnDataSource(data={
    'top': hist,
    'left': edges[:-1],
    'right': edges[1:]
})

# Create the histogram plot
p = figure(title="Histogram of Random Data",
           x_axis_label='Value', y_axis_label='Frequency')

p.quad(top='top', bottom=0, left='left', right='right', source=source,
       fill_color='navy', line_color='white', alpha=0.7)

show(p)
```

```python
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

output_notebook()
#dataset
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data for Bokeh
data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten()
}
source = ColumnDataSource(data=data)

# Create the heatmap plot
p = figure(title="Heatmap of Random Data",
           x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1),
           tools="hover", tooltips=[("Value", "@value")])

mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(data['value']), high=max(data['value']))

p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
       fill_color=mapper, line_color=None)

# Show the plot
show(p)
```
